```python
# Some imports
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set()
```

```python
# Import DBSCAN
from sklearn.cluster import DBSCAN
```

```python
# Import pandas
import pandas as pd
```

```python
# Read the csv with the snapped crash/centerline intersections
# Data was projected to WGS 1984 UTM Zone 12N with a 9 ft snapping tolerance
dtc = pd.read_csv('UtahCrashLocsRoads_Intersect09ftXY.csv')
```

```python
# Plot the entire dataset of the snapped crash/centerline intersections
axu = plt.scatter(dtc['POINT_X'], dtc['POINT_Y'], s=40)
plt.xlabel('easting')
plt.ylabel('northing')
```

```python
# Put X's and Y's of snapped crash/centerline intersections together in a numpy ndarray
stack = np.hstack((dtc[['POINT_X']].values, dtc[['POINT_Y']].values))
```

```python
# DBSCAN clustering of snapped crash/centerline intersections with eps = 10, min_samples = 3
# 1,000 point sample
dbc = DBSCAN(eps=10, min_samples=3).fit(stack)
cc = dbc.fit_predict(stack)[14000:15000]
axc = plt.scatter(stack[14000:15000, 0], stack[14000:15000, 1], c=cc, s=40, cmap='viridis')
plt.xlabel('easting')
plt.ylabel('northing')
```

```python
# DBSCAN clustering of snapped crash/centerline intersections with eps = 10, min_samples = 10
# 1,000 point sample
dbc = DBSCAN(eps=10, min_samples=10).fit(stack)
cc = dbc.fit_predict(stack)[14000:15000]
axc = plt.scatter(stack[14000:15000, 0], stack[14000:15000, 1], c=cc, s=40, cmap='viridis')
plt.xlabel('easting')
plt.ylabel('northing')
```

```python
# DBSCAN clustering of snapped crash/centerline intersections with eps = 10, min_samples = 3
# 500 point sample
dbc = DBSCAN(eps=10, min_samples=3).fit(stack)
cc = dbc.fit_predict(stack)[13700:14200]
axc = plt.scatter(stack[13700:14200, 0], stack[13700:14200, 1], c=cc, s=40, cmap='viridis')
plt.xlabel('easting')
plt.ylabel('northing')
```

```python
# DBSCAN clustering of snapped crash/centerline intersections with eps = 10, min_samples = 5
# Entire dataset, noise points removed from plot
dbc = DBSCAN(eps=10, min_samples=5).fit(stack)
cc = dbc.fit_predict(stack)
data = { 'x': stack[:,0], 'y': stack[:,1], 'label': cc}
df = pd.DataFrame(data)
df
df_filtered = df[df.label>-1]
axc = plt.scatter(df_filtered['x'], df_filtered['y'], c=df_filtered.label, s=40, cmap='viridis')
plt.xlabel('easting')
plt.ylabel('northing')
```

```python
# DBSCAN clustering of snapped crash/centerline intersections with eps = 10, min_samples = 3
# Entire dataset, noise points removed from plot
dbc = DBSCAN(eps=10, min_samples=3).fit(stack)
cc = dbc.fit_predict(stack)
data = { 'x': stack[:,0], 'y': stack[:,1], 'label': cc}
df = pd.DataFrame(data)
df
df_filtered = df[df.label>-1]
axc = plt.scatter(df_filtered['x'], df_filtered['y'], c=df_filtered.label, s=40, cmap='viridis')
plt.xlabel('easting')
plt.ylabel('northing')
```
